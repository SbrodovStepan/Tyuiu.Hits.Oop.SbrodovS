```
public abstract class Supplier
{
    public int SupplierId { get; set; }
    public string CompanyName { get; set; }
    public string ProductType { get; set; }
    public double SupplierRating { get; set; }
    public string ContactEmail { get; set; }
    public bool IsCertified { get; set; }

    public Supplier(int supplierId, string companyName, string productType, double supplierRating, string contactEmail, bool isCertified)
    {
        SupplierId = supplierId;
        CompanyName = companyName;
        ProductType = productType;
        SupplierRating = supplierRating;
        ContactEmail = contactEmail;
        IsCertified = isCertified;
    }

    public virtual string GetCompanyInfo()
    {
        return $"ID поставщика: {SupplierId}, Название компании: {CompanyName}, Тип продукции: {ProductType}, Рейтинг: {SupplierRating}, Email: {ContactEmail}, Сертификация: {IsCertified}";
    }

    public virtual string ProvideQuote()
    {
        return $"Предоставление ценового предложения для {CompanyName} в {ProductType}.";
    }

    public virtual string SubmitOrder()
    {
        return $"Заказ, отправленный в {CompanyName} для {ProductType}.";
    }

    public void RequestCertification()
    {
        if (!IsCertified)
        {
            IsCertified = true;
            Console.WriteLine($"{CompanyName} сертифицирован.");
        }
        else
        {
            Console.WriteLine($"{CompanyName} уже сертифицирован.");
        }
    }

    public void IncreaseRating(double amount)
    {
        SupplierRating += amount;
        Console.WriteLine($"Рейтинг {CompanyName} увеличен до {SupplierRating}.");
    }
}

public interface IShippable
{
    void ShipProduct();
}

public interface ITrackable
{
    void TrackOrder();
}

public class Manufacturer : Supplier, IShippable, ITrackable
{
    public int FoundedYear { get; set; }
    public int ProductionCapacity { get; set; }

    public Manufacturer(int supplierId, string companyName, string productType, double supplierRating, string contactEmail, bool isCertified, int foundedYear, int productionCapacity)
        : base(supplierId, companyName, productType, supplierRating, contactEmail, isCertified)
    {
        FoundedYear = foundedYear;
        ProductionCapacity = productionCapacity;
    }

    public override string ProvideQuote()
    {
        return $"{base.ProvideQuote()} Основанно в {FoundedYear}, Производственная мощность: {ProductionCapacity}.";
    }

    public void ShipProduct()
    {
        Console.WriteLine($"{CompanyName} отправляет продукцию.");
    }

    public void TrackOrder()
    {
        Console.WriteLine($"Заказ {CompanyName} отслеживается.");
    }
}

public class Retailer : Supplier, IShippable, ITrackable
{
    public string StoreLocation { get; set; }
    public string StoreHours { get; set; }

    public Retailer(int supplierId, string companyName, string productType, double supplierRating, string contactEmail, bool isCertified, string storeLocation, string storeHours)
        : base(supplierId, companyName, productType, supplierRating, contactEmail, isCertified)
    {
        StoreLocation = storeLocation;
        StoreHours = storeHours;
    }

    public override string SubmitOrder()
    {
        return $"{base.SubmitOrder()} Расположение магазина: {StoreLocation}, Часы работы: {StoreHours}.";
    }

    public void ShipProduct()
    {
        Console.WriteLine($"{CompanyName} отправляет заказ.");
    }

    public void TrackOrder()
    {
        Console.WriteLine($"Заказ {CompanyName} отслеживается.");
    }
}

public class Importer : Supplier, IShippable, ITrackable
{
    public string OriginCountry { get; set; }
    public bool CustomsCleared { get; set; }

    public Importer(int supplierId, string companyName, string productType, double supplierRating, string contactEmail, bool isCertified, string originCountry, bool customsCleared)
        : base(supplierId, companyName, productType, supplierRating, contactEmail, isCertified)
    {
        OriginCountry = originCountry;
        CustomsCleared = customsCleared;
    }

    public override string GetCompanyInfo()
    {
        return $"{base.GetCompanyInfo()}, Страна происхождения: {OriginCountry}, Таможня: {CustomsCleared}";
    }

    public void ShipProduct()
    {
        Console.WriteLine($"{CompanyName} отправляет товар через границу.");
    }

    public void TrackOrder()
    {
        Console.WriteLine($"Заказ {CompanyName} отслеживается после таможни.");
    }
}

Manufacturer manufacturer = new Manufacturer(1, "Компания 1", "товар 1", 4.5, "contact1@mail.com", true, 1995, 10000);
Retailer retailer = new Retailer(2, "Компания 2", "товар 2", 4.0, "contact2@mail.com", false, "Дубай", "9:00 - 18:00");
Importer importer = new Importer(3, "Компания 3", "товар 3", 3.8, "contact3@mail.com", false, "ОАЭ", false);

Console.WriteLine(manufacturer.GetCompanyInfo());
manufacturer.ShipProduct();
manufacturer.TrackOrder();

Console.WriteLine();

Console.WriteLine(retailer.GetCompanyInfo());
retailer.ShipProduct();
retailer.TrackOrder();

Console.WriteLine();

Console.WriteLine(importer.GetCompanyInfo());
importer.ShipProduct();
importer.TrackOrder();
```
